# Libraries

```python
from typing import Callable
from PIL import Image
import numpy as np
```

# Image Processor

```python
class ImageProcessor:
    """
    The ImageProcessor class contains methods for processing images using various filters.
    It applies filters to images by iterating over each pixel and applying the specified filter function.

    Methods:
    --------
    linear_filtering(image: np.ndarray, filter_function: Callable[..., np.ndarray], *args: int) -> np.ndarray
        Applies a specified linear filter to an image.
    """
    
    def linear_filtering_rgb(self, image: np.ndarray, filter_function: Callable[..., np.ndarray], *args: int) -> np.ndarray:
        """
        Applies a specified linear filter to an image.

        This method processes an image by applying a given filter function to each pixel. The filter function
        determines how each pixel's intensity should be modified. Additional arguments for the filter function can be passed
        using *args.

        Parameters:
        -----------
        image : np.ndarray
            The image to be processed, represented as a numpy array. Each element in the array represents a pixel's intensity value.
        
        filter_function : Callable[..., np.ndarray]
            The filter function to be applied to each pixel. This function should take at least one integer argument (the pixel intensity)
            and may take additional arguments depending on the filter type.
        
        *args : int
            Additional arguments to be passed to the filter function. These can be used for filters that require extra parameters (e.g., threshold values).

        Returns:
        --------
        np.ndarray
            The processed image after applying the filter, represented as a numpy array with the same shape as the input image.
        """
        
        #Código antigo
        '''        
        linear_filter = LinearFilter()
        rows_number = image.shape[0]
        cols_number = image.shape[1]
        for i in range(rows_number):
            for j in range(cols_number):
                for k in range(3):
                    image[i, j, k] = filter_function(image[i, j, k], *args)
        '''
        filtered_image = image.copy()
        
        # Apply the filter function to each channel
        for k in range(3):
            filtered_image[:, :, k] = filter_function(filtered_image[:, :, k], *args)
        
        return filtered_image
    
    
    def linear_filtering_yiq(self, image: np.ndarray, filter_function: Callable[..., np.ndarray], *args: int) -> np.ndarray:
        """
        Applies a specified filter to the Y channel of an image and converts it back to RGB.

        Parameters:
        -----------
        image : np.ndarray
            The image to be processed, represented as a numpy array in RGB format.
        
        filter_function : Callable[..., np.ndarray]
            The filter function to be applied to the Y channel.
        
        *args : int
            Additional arguments to be passed to the filter function.

        Returns:
        --------
        np.ndarray
            The processed image after applying the filter to the Y channel and converting back to RGB.
        """
        
        # Convert RGB to YIQ
        Y = image[:, :, 0] * 0.299 + image[:, :, 1] * 0.587 + image[:, :, 2] * 0.114
        I = image[:, :, 0] * 0.5959 - image[:, :, 1] * 0.2746 - image[:, :, 2] * 0.3213
        Q = image[:, :, 0] * 0.2115 - image[:, :, 1] * 0.5227 + image[:, :, 2] * 0.3112
        
        # Apply filter to the Y channel
        Y_filtered = filter_function(Y, *args)
        
        # Convert back from YIQ to RGB
        R = Y_filtered + 0.956 * I + 0.619 * Q
        G = Y_filtered - 0.272 * I - 0.647 * Q
        B = Y_filtered - 1.106 * I + 1.703 * Q
        
        # Stack channels back into an RGB image
        rgb_image = np.stack((R, G, B), axis=-1)
        
        # Clip values to ensure they are in the 0-255 range and convert to uint8
        rgb_image = np.clip(rgb_image, 0, 255).astype(np.uint8)
        
        return rgb_image
```

# Pontual Filter Class

```python
class LinearFilter:
    """
    The LinearFilter class contains methods to apply different types of linear filters
    to pixel intensity values, commonly used in digital image processing.

    Methods:
    --------
    triangular(image : np.ndarray) -> np.ndarray
        Applies a triangular filter to the pixel intensity value.
        
    binarization(image: np.ndarray, l: int) -> np.ndarray
        Applies a binarization filter to the pixel intensity value based on a threshold.
    """
    
    def triangular(self, image: np.ndarray) -> np.ndarray:
        """
        Applies a triangular filter to the entire image using NumPy, ensuring integer results.

        Parameters:
        -----------
        image : np.ndarray
            The image array to be filtered.

        Returns:
        --------
        np.ndarray
            The filtered image with integer pixel values.
        """
        
        # Lógica Sem Numpy
        '''
        if 0 <= x <= 128:
            return int((255 / 128) * x)
        elif 128 < x <= 255:
            return int(255 - (255 / 127) * (x - 128))
        else:
            return 0
        '''
        result = np.zeros_like(image)
        condition1 = (0 <= image) & (image <= 128)
        condition2 = (128 < image) & (image <= 255)
        
        # Apply operations to separate arrays
        result[condition1] = np.floor((255 / 128) * image[condition1])
        result[condition2] = np.floor(255 - (255/127) * (image[condition2] - 128))
        
        return result.astype(int)
        
    def binarization(self, image: np.ndarray, l: int) -> np.ndarray:
        """
        Applies a binarization filter to the pixel intensity value.
        If the pixel intensity value is greater than or equal to the specified threshold, it returns 255 (white);
        otherwise, it returns 0 (black).

        Parameters:
        -----------
        image : np.ndarray
            The image array to be filtered.
        l : int
            The threshold for binarization. If the pixel value is greater than or equal to this threshold, it will be set to 255.

        Returns:
        --------
        int
            The pixel intensity value after applying the binarization filter (0 or 255).
        """
        result = np.zeros_like(image)
        condition = image >= l
        result[condition] = 255
        # Clip values to ensure they're in the 0-255 range
        # image = np.clip(image, 0, 255)
        return result.astype(int)
```

```python
# Open the image file
image = Image.open('lena.tif')

display(image)
```

## RBG Filtered Image

```python
# Convert the image to a numpy array
image_array = np.asarray(image, dtype=np.uint8)

# Apply linear filters
processor = ImageProcessor()
image_output = processor.linear_filtering_rgb(image_array, LinearFilter().triangular)
#image_output = processor.linear_filtering_rgb(image_output, LinearFilter().binarization, 120)
display(Image.fromarray(image_output))
```

## YIQ Filtered Image

```python
# Convert the image to a numpy array
image_array = np.asarray(image, dtype=np.uint8)

# Apply linear filters
processor = ImageProcessor()
image_output = processor.linear_filtering_yiq(image_array, LinearFilter().triangular)
#image_output = processor.linear_filtering_yiq(image_output, LinearFilter().binarization, 120)
display(Image.fromarray(image_output))
```
